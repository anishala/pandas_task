Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
data['Income_binary']=data.income.map({'<=50K':0,'>50K':1})
data.head()
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
data['age_group'] = pd.cut(data['age'], bins=[0, 30, 60, float('inf')], labels=['Youth', 'Adult', 'Senior'])
data.head()
```
